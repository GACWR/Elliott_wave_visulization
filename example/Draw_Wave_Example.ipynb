# How to select the perfect tick?
However, the turnover point is not always perfect on different product. Hence, before select the best pattern, tweeking the hyperparameter is important. In this case we choose three different tick parameter to see the difference of the pattern.

### Select Best Parameter on Indices
Below is an example for DJI index. The result shows slightly little different between tick 3 and others. However, Using tick=3 is more reasonable for DJI index.

```python
# select the hyperparameter for best pattern
for i in [3,4,5]:
    try:
        Elloite_plot(stock_id, start_date=datetime.date.today().replace(datetime.date.today().year - time_period_year), tick=i)
    except:
        pass
```
